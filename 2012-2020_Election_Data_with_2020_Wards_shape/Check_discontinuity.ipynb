```python
import networkx as nx
import csv
import json
import pandas as pd
```

```python
#create an empty dictionary to store each node's neighbor
adj_dict = {}

for i in range(1,7079):
    node_name = 'F'+str(i)
    adj_dict[node_name] = []
    
with open('border.csv', 'r') as csvfile:
    adj = csv.reader(csvfile)
    temp = 0
    for i in adj:
        adj_dict[i[0]].append(i[1])
        adj_dict[i[1]].append(i[0])
        #print(i)
        #G2.add_edge(i[0],i[1])
        temp = temp + 1
        
for i in range(1,7079):
    node_name = 'F'+str(i)
    adj_dict[node_name] = set(adj_dict[node_name])
```

```python
assign_df = pd.read_csv('fullWisc_Wo_Adj_constraint_weighted_objective.csv')
assign_df = assign_df[assign_df['Val']==1]
```

```python
root = {}
root['d1'] = 'F1'
root['d2'] = 'F1000'
root['d3'] = 'F2000'
root['d4'] = 'F3000'
root['d5'] = 'F4000'
root['d6'] = 'F5000'
root['d7'] = 'F6000'
root['d8'] = 'F7000'
```

```python
isolated = []

for i in range(1,9):
    district = 'd' + str(i)
    G = nx.Graph()

    node_list = list(assign_df[assign_df['dc']==district]['i'])

    G.add_nodes_from(node_list)

    for i in node_list:
        for j in adj_dict[i]:
            if j in node_list:
                G.add_edge(i,j)
    
    for i in node_list:
        if not(nx.has_path(G,root[district],i)):
            isolated.append(i)
```

```python
isolated_str = 'c('

for i in isolated:
    isolated_str = isolated_str + i[1:] + ','
```
